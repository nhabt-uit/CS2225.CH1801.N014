**Chuyển đổi hình ảnh sang grayscale**

Mục tiêu chuyển sang dạng grayscale là một tấm hình màu sẽ có ba nhóm màu chính là đỏ, xanh lá, xanh dương. Khi chuyển sang grayscale thì chỉ có hai màu là trắng và đen đối với màu đen ta có giá trị là 0 và trắng là 255. Và khi xử lý ảnh 2 màu sẽ nhanh hơn xử lý ảnh nhiều màu.

**Canny edge detection**

Đối với cạnh trong hình ảnh có sự thay đổi rõ nét về cường độ, màu sắc giữa các pixel liền kề trong ảnh. Các sự thay đổi này được gọi là gradient. Hàm gradient sẽ theo dõi mọi hướng của ảnh dưới dạng một loạt pixel.

Cap22ure.JPG

HÌNH ẢNH SAU KHI NHẬN DIỆN CANNY EDGE

Kết quả cho chúng ta thấy các gradient mạnh trong hình được thể hiện bởi màu trắng. Quá trình thực hiện là dùng bộ lọc Gaussian để làm mịn ảnh và loại bỏ nhiễu sau đó tìm cường độ gradient của hình ảnh.
